```python
import numpy as np
x=np.random.seed(1)
#x
```

```python
x=np.random.randint(0,50,1000)
x
```

```python
y=x+np.random.randint(0,10,1000)
y
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('ggplot')
plt.scatter(x,y)
plt.show
```

```python
np.corrcoef(x,y)
```

```python
z=100-np.random.normal(0,10,1000)
plt.scatter(z,x)
plt.show
```

```python
np.corrcoef(z,x)
```

```python
import pandas as pd
df=pd.DataFrame({'a':np.random.randint(0,50,1000)})
df['b']=df['a']+np.random.normal(0,10,1000)
df['c']=100-df['a']+np.random.normal(0,5,1000)
df['d']=np.random.normal(0,50,1000)
df.corr()
```

```python
pd.plotting.scatter_matrix(df,figsize=(6,6))
plt.show()
```

```python
plt.matshow(df.corr())
plt.xticks(range(len(df.columns)),df.columns)
plt.yticks(range(len(df.columns)),df.columns)
plt.colorbar()
plt.show()
```
